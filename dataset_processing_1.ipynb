```python
df1 = pd.DataFrame({
    'name': ['Oyindolapo', 'Akeem', 'McLeod', 'Tara'],
    'course': ['python intro', 'python advance', None, 'AI/ML in geophysics'],
    'grade': [70, 78, None, 90]
})

df2 = pd.DataFrame({
    'name': ['Oyindolapo', 'Frank', 'McLeod', 'Tara', 'Daniel'],
    'course': ['python intro', 'python advance', 'Mixed Reality', 'AI/ML in geophysics', 'Oceanography'],
    'level': [100, 200, 300, 400, 800]
})
```

```python
merged = pd.merge(df1, df2, how = "left", on = "name", suffixes=['', "_right"])
```

```python
pd.merge(df1, df2, how="cross",  suffixes=['', "_left"] )
```

```python
for x in merged.columns:
  if merged[x].dtype == 'O':
    merged[x].fillna('No val', inplace = True)
  else:
    merged[x].fillna(0, inplace = True)
```
