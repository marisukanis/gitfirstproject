# Reading Data

Syntax:

'pd.read_csv('filepath', sep=<<defaults ','>>, header= <<defaults 0>>, names)'

- filepath: link to file
- sep: separator, default is ',' (comma delimited)
- header: the row to use as the column name(s), default to row[0]
- names: specifies the column names to use, if the file has no header row, then 'header = None' is required
- index_col: column you want to use as an index, default is 'None'

# Removing Data:

- **drop**: to remove rows and columns 'DataFrame.drop(labels, axis

  - labels: index or columns you want to drop
  - axis: 0 for rows, 1 for columns
  - columns: if you dont want to specify labels and axis for column deletion, use columns
  - inplace (bool): default is False, 'inplace = False' returns a copy. If 'inplace = True', it modifies the existing
  - errors: either ignore or raise, default is raise

```python
# dropping columns
data.drop(columns= ['poutcome'])  # returns the table without column 'poutcome'
```

```python
# dropping columns
data.drop(columns= ['poutcome'], inplace = True) # removes column 'poutcome'
```

```python
data.columns
```

```python
# dropping rows
# data[data['job'] == 'unknown']
data.drop(data[data['job'] == 'unknown'].index, inplace = True) # drops the unknown jobs
```

```python
df1 = pd.DataFrame({
   'name': ['Maris', 'Akeem', 'McLeod', 'Tara'],
   'course': ['python intro', 'python advance', None, 'AI/ML in geophysics'],
   'grade': [70, 78, None, 90] 
})

df2 = pd.DataFrame({
   'name': ['Maris', 'Frank', 'McLeod', 'Tara', 'Daniel'],
   'course': ['python intro', 'python advance', 'Mixed Reality', 'AI/ML in geophysics', 'Oceanography'],
   'level': [100, 200, 300, 400, 800] 
})
```

```python
merged = pd.merge(df1, df2, on = 'name', how = 'left', suffixes= ['', '_right']) # or 'df1.merge(df2)'
```

```python
pd.merge(df1, df2, on = 'name', how = 'right', suffixes= ['', '_left'])
```

```python
pd.merge(df1, df2, on = 'name', how = 'inner', suffixes= ['', '_left']) 
```

```python
pd.merge(df1, df2, on = 'name', how = 'outer', suffixes= ['', '_left']) 
```

```python
pd.merge(df1, df2, how = 'cross', suffixes= ['', '_left']) 
```

# Handling Missing Data:

- imputation: mean, min, max, defined function '**fillna**'
- '**dropna**': for a column, if the missing data is 50% of the population

'*na*' means any

```python
merged
```

```python
merged.isnull().sum()  # shows the count of missing values
```

```python
merged[merged['course'].isnull()]   # can filter the null values
```

```python
merged.fillna('Missing')  # replaces all null values with word 'Missing'
```

```python
merged.course.fillna('Absent')  # you can fill null values by column
```

```python
merged.course.dtype
```

```python
for x in merged.columns:
  if merged[x].dtype == 'O':
    merged[x].fillna('No value', inplace = True)
  else:
    merged[x].fillna(0, inplace = True)
```
